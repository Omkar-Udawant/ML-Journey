```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
from sklearn.linear_model import LogisticRegression
# Changed penalty='None' to penalty=None (removed quotes to use actual None value)
lor = LogisticRegression(penalty=None, solver='sag')
lor.fit(X, y)
```

```python
print(lor.coef_)
print(lor.intercept_)
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
```

```python
def gd(X, y):
    # Insert a column of 1s at index 0 for the intercept/bias term
    X = np.insert(X, 0, 1, axis=1)
    
    # Initialize weights with 1s (matching the new shape of X)
    weights = np.ones(X.shape[1])
    lr = 1  # Learning rate
    
    for i in range(4000):
        # Calculate predictions
        y_hat = sigmoid(np.dot(X, weights))
        
        # Update weights: dot product of X transposed and the error vector
        weights = weights + lr * (np.dot(X.T, (y - y_hat)) / X.shape[0])
        
    # Return coefficients (weights[1:]) and intercept (weights[0])
    return weights[1:], weights[0]

# 3. Call the function
coef_, intercept_ = gd(X, y)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
coef_,intercept_ = gd(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```
