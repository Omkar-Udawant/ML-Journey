```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.countplot(df['parental level of education'])
```

```python
sns.countplot(x="gender", data=df)
plt.title("Count of Students by Gender")
plt.show()
```

```python
df['gender'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
plt.hist(df["math score"])
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.distplot(df["math score"])
plt.title("Distribution of Math Scores")
plt.show()
```

```python
sns.histplot(df["math score"], kde=True)
plt.title("Math Score Distribution")
plt.show()
```

```python
sns.boxplot(y=df["math score"])
plt.title("Boxplot of Math Scores")
plt.show()
```

```python
df['math score'].mean()
```

```python
sns.scatterplot(x="math score", y="reading score", hue="gender", data=df)
plt.title("Math vs Reading Score by Gender")
plt.show()
```

```python
sns.scatterplot(x="math score", y="reading score", data=df)
plt.title("Math Score vs Reading Score")
plt.show()
```

```python
sns.barplot(x="gender", y="math score", data=df)
plt.title("Average Math Score by Gender")
plt.show()
```

```python
sns.boxplot(x="gender", y="math score", data=df)
plt.title("Gender vs Math Score")
plt.show()
```

```python
sns.histplot(df["math score"], bins=20, kde=True)
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.show()
```

```python
correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
```

```python


sns.clustermap(corr, annot=True, cmap="coolwarm")
plt.title("Clustered Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df[["math score","reading score","writing score"]])
plt.show()
```

```python
sns.lineplot(x="math score", y="reading score", hue="gender", data=df)
plt.title("Math vs Reading Score by Gender")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.clustermap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
pivot = pd.pivot_table(df, values="math score", index="gender", aggfunc="mean")
print(pivot)
```
